```python
def get_data(term):

    for i in range(100):
        page=i
        response_format='.json'
        url='http://api.nytimes.com/svc/search/v2/articlesearch'+response_format
        api_key='3d9980b9b07542d9b184fe5d893f480b'
        search_term=term
        begin_date='20160509'
        end_date='20160601'
        sort = 'newest'
        # fl =['headline', 'abstract','web_url']
        hl=True
        facet_field=['document_type','source']
        facet_filter=True

        query_params={
                       'begin_date': begin_date,
                       'end_date': end_date,
                      'q': search_term,
                      'sort': sort,
        #               'fl' : fl,
                      'hl' : hl,
                      'page' : page,
                      'facet_field' : facet_field,
                      'facet_filter' : facet_filter,
                      'api-key': api_key
             }

        request=requests.get(url, params=query_params)
        data=request.json()
        Trumpdata.append(data)
        time.sleep(1)
```
