```python
import json
import pandas as pd
from textblob import TextBlob
from collections import defaultdict
from collections import Counter
import matplotlib.pyplot as plt
%matplotlib inline
import datetime as Date

```

```python
with open('cleantrump.txt') as json_data:
    trump = json.load(json_data)
    json_data.close()
with open('cleanhrc.txt') as json_data:
    hrc = json.load(json_data)
    json_data.close()
with open('cleanboth.txt') as json_data:
    both = json.load(json_data)
    json_data.close()
```

```python
trump = trump + both
hrc = hrc + both
```

```python

for i in trump:
    i['date'] = i['date'].split("T")[0]

for i in hrc:
    i['date'] = i['date'].split("T")[0]
```

```python
trumpbydate = {}
for i in trump:
    date = i['date']
    text = i['full_text']
    counts = Counter(TextBlob(text).words)
    try:
        trumpbydate[date] = trumpbydate[date] + counts
    except KeyError:
        trumpbydate[date] = counts
hrcbydate = {}
for i in hrc:
    date = i['date']
    text = i['full_text']
    counts = Counter(TextBlob(text).words)
    try:
        hrcbydate[date] = hrcbydate[date] + counts
    except KeyError:
        hrcbydate[date] = counts
```

```python
print(len(trumpbydate))
print(len(hrcbydate))
```

```python
trump = pd.DataFrame.from_dict(trumpbydate, orient='index')
hrc= pd.DataFrame.from_dict(hrcbydate, orient='index')
```

```python
trump = trump.fillna(0)
hrc = hrc.fillna(0)
```

```python
trump.index = pd.to_datetime(trump.index)
hrc.index = pd.to_datetime(hrc.index)
```

```python
trumpcolors = ["#EF001D", "#F10103", "#F21A02", "#F43604", "#F55205", "#F76E07", "#F88A08","#FAA609", "FBC20B", "#FDDF0D"] 
```

```python
clintoncolors = ["#020850", "#032061", "#044073", "#066885", "#079695", "#0AA882", "#0CBA68", "#0FCB47", "#11DD1F", "#39EF14"]
```

```python
TrumpData = []
for k, c in zip(interestwords, trumpcolors):
    values = []
    for i,j in zip(trumpepoch, trump[k]):
        line = {"x": i, "y": j}
        values.append(line)
    tempdict = {}
    tempdict["key"] = "Trump: " +k
    tempdict["values"] = values
    tempdict["color"] = c
    TrumpData.append(tempdict)
```

```python
ClintonData = []
for k, c in zip(interestwords, clintoncolors):
    values = []
    for i,j in zip(hrcepoch, hrc[k]):
        line = {"x": i, "y": j}
        values.append(line)
    tempdict = {}
    tempdict["key"] = "Hillary: " +k
    tempdict["values"] = values
    tempdict["color"] = c
    ClintonData.append(tempdict)
```

```python
forviz =  TrumpData+ClintonData
```
