```python
import json
import pandas as pd
from textblob import TextBlob
```

```python
trump = pd.DataFrame(trump)
hrc= pd.DataFrame(hrc)
both = pd.DataFrame(both)
```

```python
trump['date'] = pd.to_datetime(trump.date)
hrc['date'] = pd.to_datetime(hrc.date)
both['date'] = pd.to_datetime(both.date)
```

```python
group = [trump, hrc, both]
df = pd.concat(group, ignore_index=True)
```

```python
polarity = []
subjectivity = []
for row in df.sentiment:
    polarity.append(row[0])
    subjectivity.append(row[1])
```
