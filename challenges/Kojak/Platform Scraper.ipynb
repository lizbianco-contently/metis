```python
import pandas as pd
file = 'platform_data.txt'
df = pd.read_json(file)
```

```python
df = df.drop('party', axis=1)
```

```python
party = []
for i in df.title:
    title_as_list = i.split(' ')
    partyindex = title_as_list.index('Party')
    partyname = title_as_list[(partyindex -1)]
    party.append(partyname)
    
print(party)
```

```python
df['party'] = party
```
