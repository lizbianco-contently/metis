```python
import pandas as pd
```

```python
df= pd.read_csv('mondaynighttopics.csv')
```

```python
df.shape
```

```python
df = df.drop('session', axis=1)
```

```python
lda = df['lda']
```

```python
lda_tuples = []
for i in lda:
    i = i.strip('[]').strip('()').split('), (')
    index_list = []
    for j in i:
        templist = j.split(', ')
        tup = (templist[0], templist[1])
        index_list.append(tup)
    lda_tuples.append(index_list)
        
```

```python
len(lda_tuples)
```

```python
lda_tuples[10]
```

```python
assigned_topics = []
for i in lda_tuples:
    maxtopic = max(i, key=lambda x:x[1])
    value = float(maxtopic[1])
    if value > 0.2:
        assigned_topics.append(maxtopic[0])
    else:
        assigned_topics.append('NaN')
```

```python
len(assigned_topics)
```

```python
assigned_topics[10]
```

```python
topic_by_name = []
for i in assigned_topics:
    try:
        i = int(i)
        topic = topic_title[i]
        topic_by_name.append(topic)
    except ValueError:
        topic_by_name.append('unknown')
```

```python
len(topic_by_name)
```

```python
topic_by_name[10]
```

```python
df['topic']=topic_by_name
```

```python
roster = pd.read_csv('roster114.csv')
```

```python
names = roster.title
```

```python
lastnames = []
for i in names:
    splitnames = i.split(' ')
    lastname = splitnames[-1].lower()
    if lastname == 'Jr.':
        lastnames.append(splitnames[-2].lower())
    else:
        
        lastnames.append(lastname)
```

```python
lastnames
```

```python
roster['lastnames'] = lastnames
rosterparties = roster.party
rosterstates = roster.state
```

```python
party = []
state = []
for i in df.speaker:
    try:
        index = lastnames.index(i.lower())
        party.append(rosterparties[index])
        state.append(rosterstates[index])
    except ValueError:
        party.append('not found')
        state.append('not found')
```
