```python
import nltk
import pandas as pd
import json
from nltk.tokenize import sent_tokenize, TreebankWordTokenizer, word_tokenize, wordpunct_tokenize, WhitespaceTokenizer 
```

```python
data=[]
```

```python
with open('data113.txt') as json_data:
    d = json.load(json_data)
    for i in d:
        data.append(i)
    json_data.close()
```

```python
len(data)
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
speeches = []
speakers = []       
```

```python
for i in df.speeches:
    try:
        tempspeeches = (list(i.values()))
        tempspeakers = (list(i.keys()))
        for j in tempspeeches:
            speeches.append(j)
        for k in tempspeakers:
            speakers.append(k)
                
    except AttributeError:
        pass
```

```python
len(speeches)
```

```python
len(speakers)
```

```python
sentences = []
corpus = []
for i in speeches:
    tempsentences = sent_tokenize(i)
    sentences.append(tempsentences)
    for j in tempsentences:
        corpus.append(j)
```
