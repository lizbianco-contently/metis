```python
df= df.drop('title', axis=1)
df = df.drop('in_office', axis=1)
df = df.drop('govtrack_id', axis=1)
df = df.drop('crp_id', axis=1)
df = df.drop('congresspedia_url', axis=1)
df = df.drop('official_rss', axis=1)
df = df.drop('senate_class', axis=1)
df = df.drop('birthdate', axis=1)
```

```python

df = df.drop("level_0", axis= 1)
```

```python
first = df.firstname
middle = df.middlename
last = df.lastname
suffix = df.name_suffix
```

```python
fullname = []
for i in range(0,100):
    if suffix[i] == 'Jr.':
        full = (first[i]+' '+last[i]+' '+suffix[i])
        fullname.append(full)
    else:
        full = (first[i]+' '+last[i])
        fullname.append(full)
```

```python
df['fullname'] =fullname
```
